```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
#Pregunta 1

def S(x):
    return -1.6 *x**2+1.3*x+141.3

def R(x):
    return 1.4*x**2-5.6*x+12.6

def T(x):
    return S(x)-R(x)

x = np.linspace(-100,100,2)
solucion = fsolve(T, x)
print(f'{solucion[0]:.2f}')
print(f'{solucion[1]:.2f}')
```

Como el ejercicio trata de tiempo la función va desde 0 hasta 7.8 
El valor -5.5 es un tiempo negativo que no aplica en el ejercicio

```python
#Pregunta 15 a
def I(x):
    return 50*x+0.5*x**2

def C(x):
    return 10*x+450

x = np.arange(-100, 25,0.01)
plt.plot(x, I(x), label = 'I')
plt.plot(x, C(x), label = 'C')
plt.title('Ingresos vs Costos')
plt.xlabel('Suscriptores (miles)')
plt.ylabel('Dinero (miles de pesos)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Pregunta 15 b
def I(x):
    return 50*x+0.5*x**2

def C(x):
    return 10*x+450

def U(x):
    return I(x)-C(x)
x = np.linspace(-100,20,2)
solucion = fsolve(U, x)
print(f'{solucion[0]:.2f}')
print(f'{solucion[1]:.2f}')
```
